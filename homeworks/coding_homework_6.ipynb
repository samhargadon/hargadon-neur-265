```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import files as numpy arrays
url1 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw6/acc_hits_hw6.csv'
url2 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw6/acc_fa_hw6.csv'
url3 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw6/lc_hits_hw6.csv'
url4 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw6/lc_fa_hw6.csv'

acc_hits = np.loadtxt(url1, delimiter = ',')
acc_fa = np.loadtxt(url2, delimiter = ',')
lc_hits = np.loadtxt(url3, delimiter = ',')
lc_fa = np.loadtxt(url4, delimiter = ',')
```

```python
# Print shape of each array
print(np.shape(acc_hits))
print(np.shape(acc_fa))
print(np.shape(lc_hits))
print(np.shape(lc_fa))
```

```python
# Create time variable
time = np.linspace(0, 4, 8001)
```

```python
# Determine sampling rate (samples per second)
sampling_rate = 8001/4
sampling_rate
```

```python
# Make subplots
fig, ax = plt.subplots(2,1, figsize=(12,6))

# Plot first trial of acc_hits in red and lc_hits in blue
ax[0].plot(time, acc_hits[0,:], 'r')
ax[0].plot(time, lc_hits[0,:], 'b')

ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('LFP (uV)')
ax[0].set_title('Hit Trials')


# Plot first trial of acc_fa in red and lc_fa in blue
ax[1].plot(time, acc_fa[0,:], 'r')
ax[1].plot(time, lc_fa[0,:], 'b')

ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('LFP (uV)')
ax[1].set_title('False Alarm Trials')
```

***I think the dominant oscillation in the ACC LFP might be around 3.25-3.50 Hz. For the LC, it's harder to tell but I think it would be a little higher around 4-5Hz.***

***It's difficult to say for certain from the raw LFP data whether there is a consistent relationship between the peaks and troughs of the oscillations in the two brain areas, but they seem to be consistent in the hit trials, and inversely related in the false alarms.***

```python
N = len(acc_hits[0, :])  # Assuming all trials have the same length
dt = time[1] - time[0]  # Sampling interval

# Compute coherence using FFTs
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)

plt.figure()
plt.plot(f, cohr.real)
plt.xlim([-10,50])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Coherence")
plt.title("Coherence (ACC - Hits, LC - Hits)")
plt.show()
```

***The plot of the coherence between the ACC and LC LFP during hits shows the strongest coherence in the theta band (around 4-8 Hz).***

```python
N = len(acc_hits[0, :])  # Assuming all trials have the same length
dt = time[1] - time[0]  # Sampling interval

# Compute coherence using FFTs
xf2 = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf2 = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

T = time[-1]
scale = 2 * dt**2 / T
Sxx2 = scale * (xf2 * xf2.conj()).mean(0)
Syy2 = scale * (yf2 * yf2.conj()).mean(0)
Sxy2 = scale * (xf2 * yf2.conj()).mean(0)

cohr2 = np.abs(Sxy2) / (np.sqrt(Sxx2) * np.sqrt(Syy2))

f2 = np.fft.rfftfreq(N, dt)

plt.figure()
plt.plot(f, cohr2.real)
plt.xlim([-10,50])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Coherence")
plt.title("Coherence (ACC - FA, LC - FA)")
plt.show()
```

***The frequency band with the strongest coherence appears to be similar for both hits (around 4-8 Hz in the theta band) and false alarms. This suggests that the ACC and LC might be communicating in the theta band regardless of whether the mouse responded correctly or incorrectly.***

***However, the coherence itself seems to be generally higher during hits compared to false alarms. This could indicate a stronger and more consistent coupling between the ACC and LC when the mouse makes a correct choice.***
