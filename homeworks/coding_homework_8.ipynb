```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Load data
fear_traces = np.loadtxt('https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw8/fear_conditioning_traces_hw8%20(1).csv', delimiter = ',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw8/extinction_predictors_hw8.csv', delimiter = ',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
# Separate condition and extinction traces
conditioning_traces = fear_traces[:10458, :]
extinction_traces = fear_traces[10459:, :]
```

```python
# Print trace sizes
print('Conditioning Traces Shape:', conditioning_traces.shape)
print('Extinction Traces Shape:', extinction_traces.shape)
```

```python
# Make time variables
time_conditioning = np.linspace(0, conditioning_traces.shape[0]/15, conditioning_traces.shape[0])
time_extinction = np.linspace(0, extinction_traces.shape[0]/15, extinction_traces.shape[0])
```

```python
# Make predictors variable
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = fear_traces[:,0]

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
```

```python
# Define number of ROIs
num_rois = conditioning_traces.shape[1]

# Make r_index_conditioning variable
r_index_conditioning = np.zeros(num_rois)

# Loop through each ROI in conditioning_traces
for i in range(num_rois):
  trace_temp = conditioning_traces[:, i]
  model = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model_results = model.fit()
  predicted_trace = model_results.predict()
  r_cond = pearsonr(trace_temp, predicted_trace)
  r_index_conditioning[i] = r_cond.statistic
```

```python
# Make predictors variable
tone_array2 = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array2[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array2[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array2[i] = 1
  else:
    tone_array2[i] = 0

extinction_predictors = df(data={'Intercept': np.ones_like(tone_array2), 'tone onset': tone_array2})
r_index_extinction = np.zeros(num_rois)

# Loop through each ROI in extinction_traces
for i in range(num_rois):
  trace_temp2 = extinction_traces[:, i]
  model2 = sm.GLM(trace_temp2, extinction_predictors, family=Poisson())
  model_results2 = model2.fit()
  predicted_trace2= model_results2.predict()
  r_ext = pearsonr(trace_temp2, predicted_trace2)
  r_index_extinction[i] = r_ext.statistic
```

```python
# Correlate r_index_conditioning and r_index_extinction
corr_cond_ext = pearsonr(r_index_conditioning, r_index_extinction)
print('Correlation coefficient (r):', corr_cond_ext.statistic)
```

```python
# Make scatter plot
plt.scatter(r_index_conditioning, r_index_extinction)

# Label axes
plt.xlabel('Correlation (Conditioning)')
plt.ylabel('Correlation (Extinction)')
plt.title('Correlation of ROI Correlations (Conditioning vs Extinction)')
```

***The correlation coefficient is 0.13, which is closer to 0 and suggests a weak relationship between conditioning and extinction correlations. By looking at the scatter plot, it looks like a random scatter of points, again indicating a weak or no linear relationship. This means that a cell that responds strongly to to shock and tone onset during the conditioning session is not more likely to respond strongly to tone onset during the extinction session. This also means that it is likely that the cells that encode the tone/shock memory are not the same as the cell that form the extinction memory.***
