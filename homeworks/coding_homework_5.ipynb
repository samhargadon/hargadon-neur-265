```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import files
url1 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/acc_hits_hw5.csv'
url2 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/acc_fa_hw5.csv'

acc_hits = np.loadtxt(url1, delimiter = ",")
acc_fa = np.loadtxt(url2, delimiter = ",")
```

```python
# Print shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
# Create time variable with length same as arrays with equally spaced values between 0 and 4
time = np.linspace(0, 4, 8001)
```

```python
# Determine sampling rate (samples per second)
sampling_rate = 8001 / 4
```

```python
# Plot first trial of acc_hits
fig, ax = plt.subplots(1, 2, figsize = (12,3))

ax[0].plot(time, acc_hits[0])
ax[0].set_title('First Trial of acc_hits')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('LFP')

ax[1].plot(time, acc_fa[0], 'r')
ax[1].set_title('First Trial of acc_fa')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('LFP')

plt.show()
```

```python
# Calculate and plot ERP for acc_hits
plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP for Hits')

plt.plot([2,2], [-40,40], 'k', lw=2)
plt.plot([0,4], [0,0], 'k', lw=2)
plt.show()
```

Activity after the vertical line shows there's brain activity in the ACC when they choose the correct stimuli.

```python
# Calculate and plot ERP for acc_fa
plt.plot(time, acc_hits.mean(0))
plt.plot(time, acc_fa.mean(0))

plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP for Hits and False Alarms')

plt.plot([2,2], [-40,40], 'k', lw=2)
plt.plot([0,4], [0,0], 'k', lw=2)
plt.show()
```

For the false alarms, there is no difference in brain activity after the vertical line. This shows that the ACC is not activated when they respond to the wrong stimuli proving ACC is involved with decision making.
