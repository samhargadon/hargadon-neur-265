```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import spikes left trials as numpy arrays
url1 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw7/spikes_left_trial1.csv'
url2 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw7/spikes_right_trial1.csv'
url3 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw7/spikes_left_trial2.csv'
url4 = 'https://raw.githubusercontent.com/samhargadon/hargadon-neur-265/main/data/hw7/spikes_right_trial2.csv'

spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
# Print length of each array
print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
# Determine firing rate for each trial
firing_rate_left_trial1 = 65/30
firing_rate_right_trial1 = 45/30
firing_rate_left_trial2 = 53/30
firing_rate_right_trial2 = 69/30

# Print firing rates
print('Firing rate (trial 1 left):', firing_rate_left_trial1,'Hz')
print('Firing rate (trial 1 right):',firing_rate_right_trial1, 'Hz')
print('Firing rate (trial 2 left):', firing_rate_left_trial2,'Hz')
print('Firing rate (trial 2 right):',firing_rate_right_trial2, 'Hz')
```

Interpret these results - do you think this neuron fires more before left turn trials, or right turn trials?

***This rate at which this neuron fires does not appear to be dependent on which way the rat is turning. The firing rate was higher before the left turn in trial 1, but lower in trial 2.***

```python
# Step 8

# Make two subplots
fig, ax = plt.subplots(2,1, figsize = (10,6))

# Plot spike train data from trial 1
ax[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.', label = 'Left Trial 1')
ax[0].plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.', label = 'Right Trial 1')
ax[0].set_title('Trial 1')
ax[0].set_xlabel('Time (s)')
ax[0].legend()

# Plot spike train data from trial 2
ax[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.', label = 'Left Trial 2')
ax[1].plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.', label = 'Right Trial 2')
ax[1].set_title('Trial 2')
ax[1].set_xlabel('Time (s)')
ax[1].legend()
```

```python
# Step 10

# Calculate ISI for eah trial (time between consecutive spikes)
isi_left_trial1 = np.diff(spikes_left_trial1)
isi_right_trial1 = np.diff(spikes_right_trial1)
isi_left_trial2 = np.diff(spikes_left_trial2)
isi_right_trial2 = np.diff(spikes_right_trial2)
```

```python
# Step 11

# Define the bins
bins = np.arange(0, 2, 0.01)

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(10,6))

# Create histogram for each ISI array
ax[0,0].hist(isi_left_trial1, bins)
ax[0,0].set_xlabel('ISI [s]')
ax[0,0].set_ylabel('Counts')
ax[0,0].set_title('ISI for Left Trial 1')

ax[0,1].hist(isi_right_trial1, bins)
ax[0,1].set_xlabel('ISI [s]')
ax[0,1].set_ylabel('Counts')
ax[0,1].set_title('ISI for Right Trial 1')

ax[1,0].hist(isi_left_trial2, bins)
ax[1,0].set_xlabel('ISI [s]')
ax[1,0].set_ylabel('Counts')
ax[1,0].set_title('ISI for Left Trial 2')

ax[1,1].hist(isi_right_trial2, bins)
ax[1,1].set_xlabel('ISI [s]')
ax[1,1].set_ylabel('Counts')
ax[1,1].set_title('ISI for Right Trial 2')
```

12. In a new Markdown Cell, interpret your histograms. Does this neuron tend to fire in "bursts"? Does it have higher "burstiness" during left or right turn trials? ***For left turn trials, the data seems to be more heavily concentrated on the left end of the graph than in right turn trials. This means that there were smaller ISI values before left turns, implicating more "bursting." For both graphs, the ISI values are less spread out in trial 2, indicating learning.***

```python
# Step 13

# Define time bins
time_bins = np.arange(0, 30, 1)

# Make function to calculate firing rate (trial type, time bins)
def calc_fr(spikes_trial, time_bins):
  firing_rates = np.zeros(len(time_bins)-1)
  for i in range(len(time_bins)-1):
    spk_index = (spikes_trial > time_bins[i]) & (spikes_trial < time_bins[i+1])
    true_values = spikes_trial[spk_index]
    firing_rates[i] = (len(true_values))/1
  return firing_rates

# Calculate firing rates for each trial
fr_left_trial1 = calc_fr(spikes_left_trial1, time_bins)
fr_right_trial1 = calc_fr(spikes_right_trial1, time_bins)
fr_left_trial2 = calc_fr(spikes_left_trial2, time_bins)
fr_right_trial2 = calc_fr(spikes_right_trial2, time_bins)

# Create subplots
fig, ax = plt.subplots(2,1, figsize = (12,6))

# Plot firing rates for trial 1
ax[0].plot(time_bins[:-1], fr_left_trial1, label = 'Trial 1 (Left)', color = 'r')
ax[0].plot(time_bins[:-1], fr_right_trial1, label = 'Trial 1 (Right)', color = 'b')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Firing Rate (Hz)')
ax[0].set_title('Trial 1 Firing Rates')
ax[0].legend()

# Plot firing rates for trial 2
ax[1].plot(time_bins[:-1], fr_left_trial2, label = 'Trial 2 (Left)', color = 'r')
ax[1].plot(time_bins[:-1], fr_right_trial2, label = 'Trial 2 (Right)', color = 'b')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Firing Rate (Hz)')
ax[1].set_title('Trial 2 Firing Rates')
ax[1].legend()
```
